## Sequential Train-Test Split

----------------------------------------------------

```python
def split_data(df, train_size=0.8):
    # Calculate the index at which to split the data
    split_idx = int(len(df) * train_size)
    # Split the data into train and test sets
    train = df[:split_idx]
    test = df[split_idx:]
    return train, test

# Splitting the datasets
amzn_train, amzn_test = split_data(amzn_df)
googl_train, googl_test = split_data(googl_df)
msft_train, msft_test = split_data(msft_df)
```

## Model Construction and Training

----------------------------------------------------

### Define Trading Environment

```python
class TradingEnv:
    def __init__(self, data):
        self.data = data
        self.state_space = ['Open', 'High', 'Low', 'Close', 'SMA', 'EMA', 'MACD']
        self.action_space = ['Sell', 'Hold', 'Buy']
        self.discount_factor = 0.95
        self.reset()

    def reset(self):
        self.current_step = 0
        self.total_profit = 0.0
        self.holding = False
        self.buy_price = 0.0

    def step(self, action):
        current_price = self.data['Close'].iloc[self.current_step]
        reward = 0

        if action == 2:  # Buy
            if not self.holding:
                self.buy_price = current_price
                self.holding = True

        elif action == 0:  # Sell
            if self.holding:
                reward = (current_price - self.buy_price) / (self.buy_price + 1e-8) * 100  # Avoid division by zero
                self.total_profit += reward
                self.holding = False

        self.current_step += 1
        done = self.current_step == len(self.data) - 1

        return reward, done

    def get_state(self):
        return self.data.iloc[self.current_step][self.state_space].values
```

### Construct DQN-LSTM Model

```python
class DQN_LSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(DQN_LSTM, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        h_0 = torch.zeros(1, x.size(0), hidden_dim).to(x.device)
        c_0 = torch.zeros(1, x.size(0), hidden_dim).to(x.device)
        out, _ = self.lstm(x, (h_0, c_0))
        out = self.fc(out[:, -1, :])
        return out
```

### Train Model


```python
def train_dqn_lstm(env, model, optimizer, criterion, num_episodes=5, batch_size=64, gamma=0.95, epsilon=1.0, epsilon_decay=0.995, epsilon_min=0.01):
    memory = deque(maxlen=2000)

    for episode in range(num_episodes):
        env.reset()
        state = torch.tensor([env.get_state()], dtype=torch.float32).to(device)
        total_reward = 0

        while True:
            if np.random.rand() <= epsilon:
                action = np.random.choice(len(env.action_space))
            else:
                with torch.no_grad():
                    q_values = model(state.unsqueeze(0))
                    action = torch.argmax(q_values).item()

            reward, done = env.step(action)
            total_reward += reward
            next_state = torch.tensor([env.get_state()], dtype=torch.float32).to(device)
            memory.append((state, action, reward, next_state, done))
            state = next_state

            if done:
                break

            if len(memory) > batch_size:
                batch = random.sample(memory, batch_size)
                for s, a, r, s_next, d in batch:
                    q_update = r
                    if not d:
                        q_update = (r + gamma * torch.max(model(s_next.unsqueeze(0))).item())
                    q_values = model(s.unsqueeze(0))
                    q_values[0][a] = q_update
                    optimizer.zero_grad()
                    loss = criterion(model(s.unsqueeze(0)), q_values)
                    loss.backward()
                    optimizer.step()

        if epsilon > epsilon_min:
            epsilon *= epsilon_decay

    return model
```

## Define Performance Metrics

----------------------------------------------------

```python
def evaluate_dqn_lstm(env, model):
    env.reset()
    state = torch.tensor([env.get_state()], dtype=torch.float32).to(device)
    total_profit = 0

    while True:
        with torch.no_grad():
            q_values = model(state.unsqueeze(0))
            action = torch.argmax(q_values).item()

        reward, done = env.step(action)
        total_profit += reward
        state = torch.tensor([env.get_state()], dtype=torch.float32).to(device)

        if done:
            break

    return total_profit


def cumulative_return(data):
    initial_value = data['Close'].iloc[0]
    final_value = data['Close'].iloc[-1]
    if initial_value == 0:
        return np.nan  # Avoid division by zero
    return (final_value - initial_value) / initial_value * 100

def sharpe_ratio(data, risk_free_rate=0.0):
    returns = data['Close'].pct_change().dropna()
    mean_return = returns.mean()
    std_return = returns.std()
    if std_return == 0:
        return np.nan  # Avoid division by zero
    return ((mean_return - risk_free_rate) / std_return) * 100

def maximum_drawdown(data):
    rolling_max = data['Close'].cummax()
    drawdown = (data['Close'] - rolling_max) / rolling_max
    return drawdown.min() * 100
```

## Evaluate Model

----------------------------------------------------

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
results = []

input_dim = len(['Open', 'High', 'Low', 'Close', 'SMA', 'EMA', 'MACD'])
hidden_dim = 64
output_dim = 3  # Buy, Sell, Hold
learning_rate = 0.001

datasets = {
    'AMZN': amzn_df,
    'GOOGL': googl_df,
    'MSFT': msft_df
}

for name, data in datasets.items():
    # Clean the data to remove any infinities or NaNs
    data.replace([np.inf, -np.inf], np.nan, inplace=True)
    data.dropna(inplace=True)

    env = TradingEnv(data)
    model = DQN_LSTM(input_dim, hidden_dim, output_dim).to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    criterion = nn.MSELoss()

    # Train the model
    model = train_dqn_lstm(env, model, optimizer, criterion)

    # Evaluate the model
    total_profits = evaluate_dqn_lstm(env, model)
    cr = cumulative_return(data)
    sr = sharpe_ratio(data)
    mdd = maximum_drawdown(data)

    results.append({
        'Dataset': name,
        'Total Profits ($)': total_profits,
        'Cumulative Return (%)': cr,
        'Sharpe Ratio (%)': sr,
        'Maximum Drawdown (%)': mdd
    })

results_df = pd.DataFrame(results)

# Display the results DataFrame
print(results_df)
```
